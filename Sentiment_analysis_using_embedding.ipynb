```python
docs = ['go india',
		'india india',
		'hip hip hurray',
		'jeetega bhai jeetega india jeetega',
		'bharat mata ki jai',
		'kohli kohli',
		'sachin sachin',
		'dhoni dhoni',
		'modi ji ki jai',
		'inquilab zindabad']
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts(docs)
```

```python
len(tokenizer.word_index)
```

```python
sequences = tokenizer.texts_to_sequences(docs)
sequences
```

```python
from keras.utils import pad_sequences
sequences = pad_sequences(sequences,padding='post')
sequences
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Input

vocab_size = 18   # 0..17
seq_len = 5
embed_dim = 2

model = Sequential()
model.add(Input(shape=(seq_len,)))   # ğŸ‘ˆ tells Keras the shape
model.add(Embedding(vocab_size, embed_dim))

model.summary()  # now shows (None, 5, 2) and params = 18*2 = 36
```

```python
model.compile('adam','accuracy')
```

```python
pred = model.predict(sequences)
print(pred)
```

```python
from keras.datasets import imdb
from tensorflow.keras.preprocessing.text import Tokenizer
from keras.utils import pad_sequences
from keras import Sequential
from keras.layers import Dense,SimpleRNN,Embedding,Flatten
```

```python
(X_train,y_train),(X_test,y_test) = imdb.load_data()
```

```python
X_train = pad_sequences(X_train,padding='post',maxlen=50)
X_test = pad_sequences(X_test,padding='post',maxlen=50)
```

```python
X_train.shape
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense

vocab_size = 10000
embed_dim = 2
max_len = 50   # sequence length after padding

model = Sequential([
    tf.keras.Input(shape=(max_len,)),         # ğŸ‘ˆ define input shape here
    Embedding(vocab_size, embed_dim),         # no input_length needed
    SimpleRNN(32, return_sequences=False),
    Dense(1, activation='sigmoid')
])

model.summary()
```

```python
import numpy as np

print(X_train.shape)
print("Max index in X_train:", np.max(X_train))
print("Max index in X_test :", np.max(X_test))
```

```python
vocab_size = int(max(np.max(X_train), np.max(X_test)) + 1)
print("vocab_size =", vocab_size)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
import tensorflow as tf

vocab_size = int(max(np.max(X_train), np.max(X_test)) + 1)
max_len = X_train.shape[1]   # sequence length (timesteps)

model = Sequential([
    tf.keras.Input(shape=(max_len,)),      # (batch_size, max_len)
    Embedding(vocab_size, 2),             # ğŸ‘ˆ updated vocab_size
    SimpleRNN(32),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
history = model.fit(X_train, y_train, epochs=5, validation_data=(X_test, y_test))
```
