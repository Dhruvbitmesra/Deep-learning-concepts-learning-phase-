```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
from patsy import dmatrix, bs
import statsmodels.api as sm
```

```python
df = pd.read_csv("advertising.csv")
print("Dataset shape:", df.shape)
print(df.head())
print(df.describe())
```

```python
X = df[['TV']]
y = df['Sales']
```

```python
lin_model = LinearRegression()
lin_model.fit(X, y)
y_pred_linear = lin_model.predict(X)
```

```python
r2_linear = r2_score(y, y_pred_linear)
mse_linear = mean_squared_error(y, y_pred_linear)

print("\nðŸ”¹ Linear Regression Results:")
print(f"RÂ² Score: {r2_linear:.4f}")
print(f"MSE: {mse_linear:.4f}")
print(f"Equation: Sales = {lin_model.intercept_:.2f} + {lin_model.coef_[0]:.4f} * TV")
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['TV'], y=df['Sales'], color="blue", label="Actual Data")
plt.plot(df['TV'], y_pred_linear, color="red", label="Linear Fit")
plt.text(100, 30, f"Slope: {lin_model.coef_[0]:.2f}\nIntercept: {lin_model.intercept_:.2f}",
         bbox=dict(facecolor='white', alpha=0.5))
plt.title("Linear Regression: Sales vs TV Advertising Spend")
plt.xlabel("TV Advertising Spend (in thousands)")
plt.ylabel("Sales (in thousands)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# -----------------------------
# Spline Regression
# -----------------------------
# Choose knots at 33rd, 50th, 66th percentiles
```

```python
percentiles = np.percentile(df['TV'], [33, 50, 66])
print(f"Chosen Knots: {percentiles}")
```

```python
transformed_x = bs(df['TV'], knots=percentiles, degree=3, include_intercept=True)
```

```python
spline_model = sm.OLS(y, transformed_x).fit()
y_pred_spline = spline_model.predict(transformed_x)
```

```python
r2_spline = r2_score(y, y_pred_spline)
mse_spline = mean_squared_error(y, y_pred_spline)

print("\nðŸ”¹ Spline Regression Results:")
print(f"RÂ² Score: {r2_spline:.4f}")
print(f"MSE: {mse_spline:.4f}")
print("\nSpline Model Summary:")
print(spline_model.summary())
```

```python
X_sorted = np.linspace(df['TV'].min(), df['TV'].max(), 200)
transformed_sorted = bs(X_sorted, knots=percentiles, degree=3, include_intercept=True)
y_pred_sorted = spline_model.predict(transformed_sorted)

plt.figure(figsize=(10,6))
sns.scatterplot(x=df['TV'], y=df['Sales'], color="gray", alpha=0.6, label="Actual Data")
plt.plot(df['TV'], y_pred_linear, color="red", label=f"Linear Fit (RÂ²={r2_linear:.2f})")
plt.plot(X_sorted, y_pred_sorted, color="green", label=f"Spline Fit (RÂ²={r2_spline:.2f})", linewidth=2)
for k in percentiles:
    plt.axvline(k, color="black", linestyle="--")
plt.title("Comparison: Linear vs Spline Regression on TV Spend")
plt.xlabel("TV Advertising Spend (in thousands)")
plt.ylabel("Sales (in thousands)")
plt.legend()
plt.grid(True)
plt.show()
```
